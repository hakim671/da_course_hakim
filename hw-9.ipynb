```python
import psycopg2
import sqlalchemy as db
import numpy as np
import pandas as pd
from sqlalchemy import text
from connector import connect_to
import plotly.express as px
```

Задача 1: На основе данных из таблицы track создайте столбчатую диаграмму, показывающую общую сумму минут (длительность треков) в разбивке по жанрам.

```python
with connect_to('postgres') as conn:
    df=pd.read_sql("""select * from track""", conn)
df['minute'] = df['milliseconds']/60000
df['mb'] = df['bytes']/1024/1024
px.bar(
    data_frame=df.groupby(['genre_id'],as_index=False)['minute'].sum(),
    y='minute',
    x='genre_id'
)
```

Задача 2: На основе данных из таблицы track создайте диаграмму рассеивания, показывающую соотношение минут и мегабайтов.

```python
px.scatter(
    data_frame=df,
    x='minute',
    y='mb'
)
```

Задача 3: На основе данных из таблицы invoice создайте коробчатую диаграмму, показывающую распределение дохода (поле total) в разбивке по странам.

```python
with connect_to('postgres') as conn:
    df2=pd.read_sql("""select * from invoice""", conn)
px.box(
    data_frame=df2.groupby(['billing_country'],as_index=False)['total'].sum(),
    x='billing_country',
    y='total'
)
```

Задача 4: На основе данных из таблицы invoice создайте линейчатую диаграмму, показывающую изменение дохода по месяцам за 2025 год.

```python
with connect_to('postgres') as conn:
    df3=pd.read_sql("""select extract(month from invoice_date) as month, total from invoice where extract(year from invoice_date)=2025""", conn)
px.bar(
    data_frame=df3.groupby(['month'],as_index=False)['total'].sum(),
    x='month',
    y='total',
)
```

Задача 5: На основе данных из таблицы invoice, создайте круговую диаграмму, показывающую соотношение продаж в США и в остальном мире. Т.е. ваша круговая диаграмма должна состоять лишь из двух секций: США и остальной мир

```python
with connect_to('postgres') as conn:
    df4=pd.read_sql("""select total, (case billing_country when 'USA' then 'USA' else 'others' end) as country from invoice""", conn)
px.pie(
    data_frame=df4.groupby(['country'],as_index=False)['total'].sum(),
    values='total',
    names='country'
)
```
